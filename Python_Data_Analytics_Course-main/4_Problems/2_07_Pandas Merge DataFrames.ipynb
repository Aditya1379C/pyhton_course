```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Merge US Jobs and Monthly Postings (2.7.1)

```python
# df_US = df[df['job_location'] == 'United States']
# df['job_posted_date']=df['job_posted_date'].dt.to_period('M')
# df_monthly_postings = df_US.groupby('job_posted_month').size().reset_index(name='job_postings')
# df_monthly = pd.DataFrame(df_monthly_postings)
# df_merged = df_US.merge(df_monthly, on='job_posted_month', how='right')
df['job_posted_month'] = df['job_posted_date'].dt.to_period('M')
df_us_jobs = df[df['job_country'] == 'United States']
df_monthly_postings = df_us_jobs.groupby('job_posted_month').size().reset_index(name='postings_count')
merged_df = pd.merge(df_us_jobs, df_monthly_postings, on='job_posted_month', how='right')
merged_df.head()
```

---

## 🟨 Merge Non-Remote Jobs and Salary (2.7.2)

```python
df_not_remote_jobs = df[df['job_work_from_home'] == False]
df_not_remote_salary = df_not_remote_jobs.groupby('job_title_short')['salary_year_avg'].mean()
df_not_remote_salary =pd.DataFrame(df_not_remote_salary)
df_not_remote_jobs.merge(df_not_remote_salary, on='job_title_short', how='right').head(5)
```

---

## 🟨 Merge Company Salary and Remote Count (2.7.3)

```python
df_company_salary = df.groupby('company_name')['salary_year_avg'].mean()
df_company_remote = df[df['job_work_from_home'] == True].groupby('company_name').size()
df_company_remote = pd.DataFrame(df_company_remote)
df_company_salary = pd.DataFrame(df_company_salary)
df_company_remote_merged = df_company_salary.merge(df_company_remote, on='company_name', how='inner')
df_company_remote_merged.head(5)
```
