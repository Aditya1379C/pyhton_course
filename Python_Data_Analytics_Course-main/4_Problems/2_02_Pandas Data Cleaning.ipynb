```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Remove Rows with Missing Values (2.2.1)

```python
df_filled = df.dropna(subset=['salary_year_avg'])
print('original_no_of rows:', len(df))
print('no_of rows after dropping nulls:', len(df_filled))
print('rows removed:',len(df) - len(df_filled))
```

---

## 🟨 Remove Duplicate Rows (2.2.2)

```python
df_unique = df.drop_duplicates(subset=['job_location'])
print('original_no_of rows:', len(df))
print('no_of rows after dropping duplicates:', len(df_unique))
print('rows removed:',len(df) - len(df_unique))
```
