```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Convert Date to String (2.10.1)

```python
df['job_posted_date_str'] = df['job_posted_date'].apply(lambda x: x.strftime('%Y-%m-%d')).copy()
df.head(5)
```

---

## 🟨 Days Since Posted (2.10.2)

```python
# df['days_since_posted']= df['job_posted_date'].apply(lambda x: (pd.Timestamp.now() - x).days).copy()
from datetime import datetime
current_date = datetime.now()
df['days_since_posted'] = df['job_posted_date'].apply(lambda date: (current_date - date).days)

df.head(5)
```

---

## 🟥 Salary Category (2.10.3)

```python
df_filtered = df[df['salary_year_avg'].notnull()].copy()
df_filtered['salary_category'] = df_filtered['salary_year_avg'].apply(lambda x: 'High' if x >100000 else 'medium' if 60000<x<100000 else 'low' if x <60000 else 'unknown')
df_filtered[['salary_year_avg', 'salary_category']].head(5)
```
